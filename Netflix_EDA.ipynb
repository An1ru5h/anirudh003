```python
round(missing_count/df.shape[0]*100,2)
```

```python
drop_date_added=df.dropna(subset=['date_added'],axis=0, inplace=True)
df.shape
```

```python
round(df.isnull().sum()/df.shape[0]*100,2)
```

```python
df.dropna(subset=['rating','duration'],axis=0,inplace=True)
round(df.isnull().sum()/df.shape[0]*100,2)
```

```python
df.dropna(subset=['date_added'],axis=0,inplace=True)
round(df.isnull().sum()/df.shape[0]*100)
```

```python
df["country"].replace(np.NaN,"Unknown",inplace= True)
```

```python
df["cast"].replace(np.NaN,"No Cast",inplace= True)
```

```python
df.cast.value_counts().head(10)
```

```python
df["director"].replace(np.NaN,"No Director",inplace= True)
```

```python
round(df.isnull().sum()/df.shape[0]*100,2)
```

```python
df['title']
```

```python
movies_df=df.loc[(df['type']=='Movie')]
movies_df.head(5)
```

```python
movies_df.duration=movies_df.duration.apply(lambda x:x.replace("min","") if 'min'in x else x)
```

```python
tvshows_df=df.loc[(df['type']=='TV Show')]
tvshows_df.head(5)
```

```python
movies_df["duration"] = movies_df["duration"].apply(lambda x: pd.to_numeric(x, errors='ignore'))
movies_df.info()
```

```python
shortest_movies = movies_df.loc[(movies_df['duration'] == np.min(movies_df.duration))]
shortest_movies
```

```python
longest_movies = movies_df.loc[(movies_df['duration'] == np.max(movies_df.duration))]
longest_movies
```

```python
tvshows_df['duration'] = tvshows_df.apply(lambda row: row['duration'].replace('Season', '') if 'Season' in row['duration'] else row['duration'], axis=1)
tvshows_df.head(2)
```

```python
tvshows_df['duration'] = tvshows_df.apply(lambda row: row['duration'].replace('s', '') if 's' in row['duration'] else row['duration'], axis=1)
tvshows_df.head(2)
```

```python
tvshows_df.describe()
```

```python
tvshows_df.loc[:,['duration']]=tvshows_df.loc[:,['duration']].apply(lambda x:x.astype('int64',errors='ignore'))
tvshows_df.describe()
```

```python
tvshows_df.duration.value_counts().tail(10)
```

```python
#Shows with highest no of seasons

longest_shows=tvshows_df.loc[(tvshows_df.duration)>=13]
longest_shows
```

```python
longest_shows.rating.value_counts()
```

```python
netflix_date=df[['date_added']].dropna()
netflix_date['year']=netflix_date['date_added'].apply(lambda x:x.split(',')[-1])
netflix_date['month']=netflix_date['date_added'].apply(lambda x:x.lstrip().split(' ')[0])

month_order=['January','February','March','April','May','June','July','August','September','October','November','December']
new_df=netflix_date.groupby('year')['month'].value_counts().unstack().fillna(0)[month_order]
plt.figure(figsize=(10,7))
#Heatmap
plt.pcolor(new_df,cmap='Blues',edgecolor="white",linewidths=2)
plt.xticks(np.arange(0.5,len(new_df.columns),1),new_df.columns,fontsize=7)
plt
plt.yticks(np.arange(0.5,len(new_df.index),1),new_df.index,fontsize=7)

plt.title('Months V/S Netflix Contents Updates',fontsize=12)
cbar=plt.colorbar()

cbar.ax.tick_params(labelsize=1)
cbar.ax.minorticks_on()
plt.show()
```
